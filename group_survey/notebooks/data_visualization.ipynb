```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
labels = ['Yes', 'No']
grade_avgs = [7.44, 6]
sleep_avgs = [7.33, 6.67]

fig, axes = plt.subplots(1, 2, figsize = (10,4), facecolor='lightgrey')

axes[0].bar([1, 2], grade_avgs, color='green', alpha=0.7)
axes[0].set_xticks([1, 2])
axes[0].set_xticklabels(['Yes', 'No'])
axes[0].set_xlabel('Same Grade')
axes[0].set_ylabel('Average Roommate Rating')

axes[1].bar([1, 2], sleep_avgs, color='pink', alpha=0.7)
axes[1].set_xticks([1, 2])
axes[1].set_xticklabels(['Yes', 'No'])
axes[1].set_xlabel('Same Sleep Schedule')
axes[1].set_ylabel('Average Roommate Rating')

fig.tight_layout()
fig.show()
fig.savefig("roommate_rating_comparison.svg")
```

```python
pie_labels = ['Sports', 'Arts', 'Academics', 'None']
pie_data = [8, 2, 6, 5]
fig, ax = plt.subplots(figsize = (6,4), facecolor='lightgrey')
ax.pie(pie_data, autopct='%1.1f%%', startangle=90)
ax.legend(pie_labels, title='Roommate Shared Interests', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

fig.tight_layout()
fig.show()
fig.savefig("roommate_shared_interests.svg")
```

```python
hist_data = [4, 5, 5, 6, 6, 6, 7, 7, 7, 8, 8, 8, 8, 8, 8, 8, 8, 8, 9, 9, 9]

fig, ax = plt.subplots(figsize = (6,4), facecolor='lightgrey')
ax.hist(hist_data, bins=6, color='blue', edgecolor='black', alpha=0.7)
ax.set_xlabel('Roommate Rating')
ax.set_ylabel('Frequency')

fig.tight_layout()
fig.show()
fig.savefig("roommate_rating_histogram.svg")
```
