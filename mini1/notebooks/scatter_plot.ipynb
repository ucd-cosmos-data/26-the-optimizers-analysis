```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load data
df = pd.read_csv('../data/interim/depression_precipitation_ca.csv')

# auto-detect columns
dep_col = next((c for c in df.columns if 'depress' in c.lower()), None)
prec_col = next((c for c in df.columns if 'precip' in c.lower() or 'rain' in c.lower()), None)

if prec_col is None and "Value" in df.columns:
    prec_col = "Value"

if dep_col is None or prec_col is None:
    raise ValueError(f"Couldn't detect columns. Available columns: {list(df.columns)}")

# prepare data
data = df[[dep_col, prec_col]].dropna()
x = data[prec_col].astype(float).values
y = data[dep_col].astype(float).values

# scatter + linear fit + correlation
m, b = np.polyfit(x, y, 1)
r = np.corrcoef(x, y)[0, 1]

fig = plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.7)
xs = np.linspace(x.min(), x.max(), 200)
plt.plot(xs, m*xs + b, color='red', lw=1.5, label=f'Fit: y={m:.3f}x+{b:.3f}')
plt.plot(label=f'R value: {r:.3f}')
plt.xlabel("Precipitation (mm)")
plt.ylabel("Depression")
plt.legend()
plt.grid(True)

plt.text(
    0.05, 0.95,
    f"r = {r:.3f}",
    transform=plt.gca().transAxes,
    verticalalignment="top"
)

plt.show()
```
